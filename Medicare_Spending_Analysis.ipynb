# Data Transformation

```python
medicare_data_2022.columns = ['Facility Name', 'Facility ID', 'State', 'Period', 'Claim Type', 'Avg Spending Per Episode (Hospital)', 'Avg Spending Per Episode (State)', 'Avg Spending Per Episode (Nation)', 'Percent of Spending (Hospital)', 'Percent of Spending (State)', 'Percent of Spending (Nation)', 'Measure Start Date', 'Measure End Date']
medicare_data_2015.columns = ['Facility Name', 'Facility ID', 'State', 'Period', 'Claim Type', 'Avg Spending Per Episode (Hospital)', 'Avg Spending Per Episode (State)', 'Avg Spending Per Episode (Nation)', 'Percent of Spending (Hospital)', 'Percent of Spending (State)', 'Percent of Spending (Nation)', 'Measure Start Date', 'Measure End Date'] 
```

Concatenating both the dataframes into a single dataframe (vertical merge)

```python
medicare_data = pd.concat([medicare_data_2015, medicare_data_2022])
```

```python
medicare_data = medicare_data.drop(columns=['Measure Start Date','Measure End Date'])
```
